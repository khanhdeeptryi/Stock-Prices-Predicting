```python
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima import auto_arima
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/VIX_Cleaned.csv')
```

```python
# Tiền xử lý: định dạng thời gian và đặt index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
ts_series = df['Close'].asfreq('B').fillna(method='ffill')  # dùng business day + ffill nếu thiếu
```

```python
# 3. TỰ ĐỘNG TÌM SIÊU THAM SỐ
model = auto_arima(ts_series, seasonal=False, trace=True, suppress_warnings=True)
print(model.summary())
```

Tìm siêu tham số với auto_arima

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
plot_acf(ts_series.diff().dropna(), ax=axes[0])
axes[0].set_title('ACF (after 1st differencing)')
plot_pacf(ts_series.diff().dropna(), ax=axes[1])
axes[1].set_title('PACF (after 1st differencing)')
plt.tight_layout()
plt.show()
```
